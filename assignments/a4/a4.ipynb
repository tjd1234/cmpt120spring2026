You've been hired Amazing.com to analyze their customer's shopping habits. 

## Products
Products are listed in [products.txt](products.txt). Each line has the form
"`code name`", where `code` is a unique integer that identifies the product, and
`name` is the name of the product. For example, one line might be:

```
70 Ultra Goldfish Cannon
```

70 is the code, and everything after that is the name of the product, "Ultra
Goldfish Cannon" in this case.

Conveniently, if there are $n$ products in
the file, then the codes are $0, 1, \ldots, n-1$.

## Baskets
Whenever a customer buys something, their purchase is recorded in a **basket**,
which is a comma-separated list of product codes. For example, here's a basket
for a customer who bought 4 items:

```
9, 121, 187, 9
```

According to [products.txt](products.txt), these codes refer too:

```
Crystal Stabilizer, Magnetic Chunker, Brine-soaked Launcher, Crystal Stabilizer
```

Note that baskets can have repeat items, and the order doesn't matter.

The file [baskets.txt](baskets.txt) contains one basket per line, and there is
always at least one item in each basket (so no empty lines). The numbers
correspond to the product codes in [products.txt](products.txt). You can
(unrealistically!) assume everything is formatted correctly and all the codes
are of a product in [products.txt](products.txt).

## Tasks

Your job is to write a program that reads both [products.txt](products.txt) and
[baskets.txt](baskets.txt), and then prints these answers:

1. **The total number of products in [products.txt](products.txt)**.
2. **The total number of baskets in [baskets.txt](baskets.txt)**. 
3. **The average number of items per basket in [baskets.txt](baskets.txt)**.
   Duplicates count, e.g. the basket `1, 2, 1` has 3 items.
4. **The percentage of baskets in [baskets.txt](baskets.txt) that have one or
   more repeated items**.
5. **The top 3 most frequent products in [baskets.txt](baskets.txt)**.
6. **The pair of products that most frequently occur together in the same basket
   in [baskets.txt](baskets.txt)**. If more than one pair is tied for most
   frequent, then list all the pairs.

When calculating which product pairs occur together in a basket, make sure to
first remove any duplicates, and be careful not to count the same pair more than
once. For example, if the basket is `1, 5, 2, 1, 5`, then removing duplicates
gives `1, 2, 5`, and the product pairs are `1, 2`, `1, 5`, and `2, 5`. The order
of the pairs doesn't matter, so, for example,`1, 2` and `2, 1` count as the same
pair. Baskets with only 1 item don't contribute any pairs.
