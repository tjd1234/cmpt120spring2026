**Misconception to address:** Some students think `home == away` checks if they are the *same object* (i.e., aliases). They are not — they are two separate lists that happen to be equal. Use `home is away` to illustrate identity vs. equality if the class is ready for it.

````python
history = []

# 1. Visit 'google.com'

# 2. Visit 'reddit.com'

# 3. Visit 'python.org'

# 4. Visit 'docs.python.org'

# 5. Press Back

# 6. Press Back again

# 7. Visit 'wikipedia.org'

# Print current page
````

**Teacher Notes — Problem 4**

## Problem 5 — Statistics on a Grade List

A teacher has these quiz scores:

```python
scores = [78, 92, 85, 61, 74, 88, 95, 70, 83, 91]
```

Using only `len`, `min`, `max`, `sum`, `in`, and `not in` (no loops yet!), compute and print:

1. The number of scores
2. The highest score
3. The lowest score
4. The total points
5. The class average (to 1 decimal place)
6. Whether anyone scored a perfect `100`
7. Whether anyone scored below `60`

**Bonus:** The range (highest minus lowest).

````python
scores = [78, 92, 85, 61, 74, 88, 95, 70, 83, 91]

# 1. Number of scores

# 2. Highest score

# 3. Lowest score

# 4. Total points

# 5. Class average (1 decimal)

# 6. Anyone scored 100?

# 7. Anyone below 60?

# Bonus: range
````

````python
# SOLUTION — Problem 5

scores = [78, 92, 85, 61, 74, 88, 95, 70, 83, 91]

print(len(scores))                               # 10
print(max(scores))                               # 95
print(min(scores))                               # 61
print(sum(scores))                               # 817
print(round(sum(scores) / len(scores), 1))       # 81.7
print(100 in scores)                             # False
print(60 not in scores)                          # True  (no one scored exactly 60)
# Note: 'in' checks for exact match, not ranges!

# Bonus
print(max(scores) - min(scores))                 # 34
````

**Teacher Notes — Problem 5**

The most important teaching point here is question 7: `in` checks for **exact membership**, not a range comparison. Students will often try `60 not in scores` thinking it means "no score below 60", but the list does not contain the value `60` exactly — there is a `61`. This is a great opportunity to clarify that `in` is not `<` or `>`.

For checking if any score is below 60, students would need a loop (which comes later). If a student asks "but how *do* we check that?", praise them for the great question and promise it comes soon.

**Average:** `sum(scores) / len(scores)` = `81.7`. Using `round(..., 1)` is a bonus concept — `f'{avg:.1f}'` is an alternative.

**Common mistake:** `sum(scores / len(scores))` — students try to divide the list by a number, which doesn't work.

## Problem 6 — Strings and Lists: Processing CSV Data

A data file has a header row stored as a string:

```python
header = 'name,  age,  city,  score'
```

Do all of the following:

1. Split `header` into a list of column names using `split(',')`.
2. Notice the extra spaces — strip each item so you get clean names like `'name'`, `'age'`, etc. Use a `for` loop with `strip()`.
3. Print the cleaned list.
4. Join the cleaned names back into a single string separated by `' | '`.
5. Convert the string `'Python'` into a list of characters, reverse it, and join back into a string.

````python
header = 'name,  age,  city,  score'

# 1. Split on ','

# 2. Strip whitespace from each item

# 3. Print cleaned list

# 4. Join with ' | '

# 5. Reverse 'Python' using list operations
````

````python
# SOLUTION — Problem 6

header = 'name,  age,  city,  score'

# 1. Split on ','
cols = header.split(',')
print(cols)  # ['name', '  age', '  city', '  score']

# 2. Strip each item in-place
for i in range(len(cols)):
    cols[i] = cols[i].strip()

# 3. Cleaned list
print(cols)  # ['name', 'age', 'city', 'score']

# 4. Join with ' | '
print(' | '.join(cols))  # 'name | age | city | score'

# 5. Reverse 'Python'
letters = list('Python')
letters.reverse()
print(''.join(letters))  # 'nohtyP'
````

**Teacher Notes — Problem 6**

Step 2 previews the "strip_list" pattern from the lecture. Key points:
- `for i in range(len(cols))` is needed here because we're **modifying** the list by index. A plain `for item in cols` would create a new value but not update the list.
- Students may try `cols[i] = cols[i].strip()` outside a loop (applying it once to the list) — show why that doesn't work.

Step 5 illustrates `list()` + `reverse()` + `join()` as a string-reversal idiom. Reinforce that these are all in-place operations on the list.

**Alternative for step 5:** `'Python'[::-1]` (step slicing on a string) works too, but uses concepts not yet covered.

````python
# SOLUTION — Problem 7

playlist = ['Jazz', 'Pop', 'Rock', 'Classical', 'Pop', 'Jazz']
print('Start:   ', playlist)

# 1. Remove first 'Pop'  → ['Jazz', 'Rock', 'Classical', 'Pop', 'Jazz']
playlist.remove('Pop')
print('Step 1:  ', playlist)

# 2. Insert 'Blues' at index 2  → ['Jazz', 'Rock', 'Blues', 'Classical', 'Pop', 'Jazz']
playlist.insert(2, 'Blues')
print('Step 2:  ', playlist)

# 3. Pop index 4  → removed = 'Pop';  ['Jazz', 'Rock', 'Blues', 'Classical', 'Jazz']
removed = playlist.pop(4)
print(f'Step 3:   popped "{removed}"', playlist)

# 4. Append 'Folk'  → ['Jazz', 'Rock', 'Blues', 'Classical', 'Jazz', 'Folk']
playlist.append('Folk')
print('Step 4:  ', playlist)

# 5. Sort  → ['Blues', 'Classical', 'Folk', 'Jazz', 'Jazz', 'Rock']
playlist.sort()
print('Step 5:  ', playlist)

# 6. Reverse  → ['Rock', 'Jazz', 'Jazz', 'Folk', 'Classical', 'Blues']
playlist.reverse()
print('Final:   ', playlist)
````

**Teacher Notes — Problem 7**

## Problem 9 — Reading and Processing a File

The file `words.txt` in this folder contains a large list of English words, one per line.

Write code to:

1. Read `words.txt` into a list called `words` (one word per element, no blank entries).
2. Print how many words are in the file.
3. Print the first 5 words and the last 5 words.
4. Sort the list and print the first 5 alphabetically and the last 5 alphabetically.
5. Find and print the **shortest** and **longest** words by character count.

*Hint for step 1:* After splitting on `'\n'`, some empty strings may appear — use a list comprehension or loop to remove them, or check `if word != ''`.

````python
# 1. Read words.txt into a list

# 2. Number of words

# 3. First 5 and last 5 words (unsorted)

# 4. Sort, then first 5 and last 5 alphabetically

# 5. Shortest and longest word
````

````python
# SOLUTION — Problem 9

# 1. Read file — filter out any empty strings from trailing newlines
all_lines = open('words.txt').read().split('\n')
words = [w for w in all_lines if w != '']
# Alternative without list comprehension:
# words = []
# for w in all_lines:
#     if w != '':
#         words.append(w)

# 2. Word count
print(f'Total words: {len(words)}')

# 3. First 5 and last 5
print('First 5:', words[:5])
print('Last 5: ', words[-5:])

# 4. Sort
words.sort()
print('First 5 alphabetically:', words[:5])
print('Last 5 alphabetically: ', words[-5:])

# 5. Shortest and longest
shortest = min(words, key=len)
longest  = max(words, key=len)
print(f'Shortest: "{shortest}" ({len(shortest)} chars)')
print(f'Longest:  "{longest}" ({len(longest)} chars)')
````

**Teacher Notes — Problem 9**

## Problem 10 — Challenge: Word Length Analysis

Using `words.txt`, answer the following questions. Write a **separate block of code** for each question.

**A.** How many words in `words.txt` are exactly **5 letters** long?

**B.** Build a sorted list of all words that are **10 or more letters long**. Print the first 10 of them.

**C.** Find all words that **start and end with the same letter** (e.g. `'level'`, `'civic'`). How many are there? Print any 10 of them.

**D.** Find all words where each letter appears **only once** (no repeated characters). How many such words are there that are **at least 8 letters long**? Print any 5.

*Hint for D:* For a word `w`, you can check `len(w) == len(list(w))` ... but that's not quite right. Think about what makes letters unique. (Hint: `set`.)

````python
# Load words first
words = [w for w in open('words.txt').read().split('\n') if w != '']

# A. Words exactly 5 letters long

# B. Words 10+ letters, sorted

# C. Words that start and end with the same letter

# D. Words of 8+ letters with no repeated characters
````

````python
# SOLUTION — Problem 10

words = [w for w in open('words.txt').read().split('\n') if w != '']

# A. Exactly 5 letters
five_letter = []
for w in words:
    if len(w) == 5:
        five_letter.append(w)
print(f'A. Words with exactly 5 letters: {len(five_letter)}')

# B. 10+ letters, sorted
long_words = []
for w in words:
    if len(w) >= 10:
        long_words.append(w)
long_words.sort()
print(f'B. Words with 10+ letters: {len(long_words)}')
print('   First 10:', long_words[:10])

# C. Start and end with same letter
same_ends = []
for w in words:
    if len(w) >= 2 and w[0] == w[-1]:
        same_ends.append(w)
print(f'C. Same first & last letter: {len(same_ends)}')
print('   Sample:', same_ends[:10])

# D. 8+ letters, no repeated characters
# set() removes duplicates; if len(set(w)) == len(w), all chars unique
unique_long = []
for w in words:
    if len(w) >= 8 and len(set(w)) == len(w):
        unique_long.append(w)
print(f'D. 8+ letters, no repeats: {len(unique_long)}')
print('   Sample:', unique_long[:5])
````

**Teacher Notes — Problem 10**

This is a substantial challenge that combines **all** the major list concepts. Expect only the strongest students to complete all four parts independently.

**Part A** is accessible to all — just a loop with an `if len(w) == 5` check.

**Part B** adds sorting. Students should call `.sort()` on their collected list.

**Part C** introduces `w[0] == w[-1]`. Some students forget to check `len(w) >= 2` (a one-letter word vacuously satisfies this, depending on whether you want to include it). Either answer is defensible; ask students to justify their choice.

**Part D** is the real challenge. Key idea: `set(w)` removes duplicate characters, so `len(set(w)) == len(w)` is True only if all characters are unique. Students who haven't seen `set` will need a hint; accept alternative solutions using loops:
```python

for ch in w:
    if w.count(ch) > 1:
        all_unique = False
        break
```

**Pedagogical note:** Problems C and D get students thinking about *string indexing* (`w[0]`, `w[-1]`) applied to words in a list — a nice synthesis of Chapter 8 (strings) and Chapter 9 (lists).
