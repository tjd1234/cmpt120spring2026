## Problem 5 — Statistics on a Grade List

A teacher has these quiz scores:

```python
scores = [78, 92, 85, 61, 74, 88, 95, 70, 83, 91]
```

Using only `len`, `min`, `max`, `sum`, `in`, and `not in` (no loops yet!), compute and print:

1. The number of scores
2. The highest score
3. The lowest score
4. The total points
5. The class average (to 1 decimal place)
6. Whether anyone scored a perfect `100`
7. Whether anyone scored below `60`

**Bonus:** The range (highest minus lowest).

````python
scores = [78, 92, 85, 61, 74, 88, 95, 70, 83, 91]

# 1. Number of scores
print('number', len(scores))

# 2. Highest score
print('highest', max(scores))

# 3. Lowest score
print('lowest', min(scores))

# 4. Total points
print('total', sum(scores))

# 5. Class average (1 decimal)
average = sum(scores) / len(scores)
print('average', average)

# 6. Anyone scored 100?
if 100 in scores:
    print('someone scored 100')
else:
    print('no one scored 100')

# 7. Anyone below 60?
if min(scores) < 60:
    print('someone score < 60')
else:
    print('no one scored < 60')

# Bonus: range
````

## Problem 6 — Strings and Lists: Processing CSV Data

A data file has a header row stored as a string:

```python
header = 'name,  age,  city,  score'
```

Do all of the following:

1. Split `header` into a list of column names using `split(',')`.
2. Notice the extra spaces — strip each item so you get clean names like `'name'`, `'age'`, etc. Use a `for` loop with `strip()`.
3. Print the cleaned list.
4. Join the cleaned names back into a single string separated by `' | '`.
5. Convert the string `'Python'` into a list of characters, reverse it, and join back into a string.

````python
header = 'name, age,  city,score\n'

# 1. Split on ','
# print(list(header)) # wrong: splits into individual characters
data = header.split(',')
print(data)

# 2. Strip whitespace from each item
# for s in data:
#     s = s.strip()
for i in range(len(data)):
    data[i] = data[i].strip()

# 3. Print cleaned list
print(data)

# 4. Join with ' | '
line = ' | '.join(data)
# line = line + '\n' + '-' * len(line)
print(line)

# 5. Reverse 'Python' using list operations
rdata = data[:]
rdata.reverse()
print(rdata)
line = ' | '.join(rdata)
print(line)
````

**Part B.** Write a function `clamp_list(lst, low, high)` that modifies `lst` **in-place** so that every value below `low` is replaced with `low`, and every value above `high` is replaced with `high`. The function should return nothing.

Example:
```python
data = [3, -1, 15, 7, 0, 22]
clamp_list(data, 0, 10)
print(data)  # [3, 0, 10, 7, 0, 10]
```

````python
# Part A — run and check your prediction
# x = [10, 20, 30]
# y = x
# y.append(40)
# print(x)
# print(y)
# print(x == y) 

# Part B — write clamp_list
def clamp_list(lst, low, high):
    """Clamp every value in lst to the range [low, high] in-place."""
    for i in range(len(lst)):
        if lst[i] < low:
            lst[i] = low
        elif lst[i] > high:
            lst[i] = high

# Test
data = [3, -1, 15, 7, 0, 22]
clamp_list(data, 0, 10)
print(data)  # should be [3, 0, 10, 7, 0, 10]
````
