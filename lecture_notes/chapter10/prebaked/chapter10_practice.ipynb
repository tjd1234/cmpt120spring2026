## Problem 2 — Building and Modifying a Dictionary

**Concepts:** empty dict, adding pairs, updating values, overwriting keys

Start with an empty dictionary called `scores`. Then:
1. Add these quiz scores one at a time: `Alice -> 88`, `Bob -> 72`, `Carol -> 95`.
2. Alice retook the quiz and scored 91. Update her score.
3. Print the final dictionary.
4. Print the number of students.
5. Does updating Alice's score change the number of entries? Explain why or why not.

````python
# Your solution here
````

````python
# === SAMPLE SOLUTION ===
scores = {}

# 1. Add entries
scores['Alice'] = 88
scores['Bob']   = 72
scores['Carol'] = 95

# 2. Update Alice
scores['Alice'] = 91

# 3. Print final dict
print(scores)        # {'Alice': 91, 'Bob': 72, 'Carol': 95}

# 4. Count
print(len(scores))   # 3

# 5. No -- updating an existing key REPLACES its value, not adds a new entry.
#    Keys are unique, so there is still just one 'Alice' entry.
````

**Teacher notes — Problem 2:**
- Key insight: assigning to an existing key updates it, it does not duplicate it.
- Students sometimes expect two Alice entries. Use this to reinforce key uniqueness.
- Discussion: what would a list look like for the same data? What are the trade-offs?

## Problem 3 — Searching with `in`

**Concepts:** `in` on keys vs. values, safe lookup pattern

Use: `scores = {'Alice': 91, 'Bob': 72, 'Carol': 95}`

1. Check whether `'Carol'` is a key in `scores`.
2. Check whether `'Dave'` is a key.
3. Check whether the score `72` appears in the values.
4. Write `safe_lookup(d, key)` that returns the value if the key exists, or `'Not found'` otherwise.
5. Test: `safe_lookup(scores, 'Bob')` should return `72`, and `safe_lookup(scores, 'Eve')` should return `'Not found'`.

````python
# Your solution here
scores = {'Alice': 91, 'Bob': 72, 'Carol': 95}
````

````python
# === SAMPLE SOLUTION ===
scores = {'Alice': 91, 'Bob': 72, 'Carol': 95}

# 1-3
print('Carol' in scores)           # True
print('Dave'  in scores)           # False
print(72 in scores.values())       # True

# 4-5
def safe_lookup(d, key):
    if key in d:
        return d[key]
    else:
        return 'Not found'

print(safe_lookup(scores, 'Bob'))  # 72
print(safe_lookup(scores, 'Eve'))  # Not found
````

**Teacher notes — Problem 3:**
- `safe_lookup` is a classic pattern. Students will use it constantly.
- Note: `72 in scores` (without `.values()`) checks keys, not values -- a common mistake.
- Advanced: why is `k in d` faster than `v in d.values()`? (Hashing vs. linear scan.)

## Problem 4 — Counting: Vote Tally

**Concepts:** counting pattern, building a frequency dict from a file

The file `votes.txt` contains one name per line -- each line is a vote for that person.

1. Read `votes.txt` and build a dictionary `tally` mapping each name to their vote count.
2. Print the final tally.
3. Print the name of the candidate with the **most** votes.

*Hint: use `if name in tally: tally[name] += 1  else: tally[name] = 1`*

````python
# Your solution here
````

````python
# === SAMPLE SOLUTION ===

# 1. Build tally
tally = {}
for line in open('votes.txt'):
    name = line.strip()
    if name in tally:
        tally[name] += 1
    else:
        tally[name] = 1

# 2. Print
print(tally)

# 3. Find winner
winner = ''
best   = 0
for name in tally:
    if tally[name] > best:
        best   = tally[name]
        winner = name

print(f'Winner: {winner} with {best} votes')
````

**Teacher notes — Problem 4:**
- The counting pattern `if k in d: d[k]+=1 else: d[k]=1` is the most important pattern of this chapter.
- votes.txt has 10,000 votes. Mention the O(1) per update advantage of dictionaries.
- Advanced shorthand: `max(tally, key=tally.get)` finds the winner in one line.
- Common mistake: students forget `.strip()` and the newline becomes part of the name.

## Problem 5 — Fast Membership Testing

**Concepts:** dictionary as a fast set, comparing list vs. dict lookup speed

`words.txt` contains over 100,000 English words.

1. Load all words into a **list** called `word_list`.
2. Load all words into a **dictionary** called `word_dict` (map each word to `1`).
3. Write `is_english_word_list(w)` that checks `w in word_list`.
4. Write `is_english_word_dict(w)` that checks `w in word_dict`.
5. Test both with `'python'` (valid) and `'xqzyw'` (invalid).
6. (Optional) Use the `time` module to measure which approach is faster.

````python
# Your solution here
````

````python
# === SAMPLE SOLUTION ===
import time

word_list = open('words.txt').read().split('\n')

word_dict = {}
for w in word_list:
    word_dict[w] = 1

def is_english_word_list(w):
    return w in word_list

def is_english_word_dict(w):
    return w in word_dict

print(is_english_word_list('python'), is_english_word_dict('python'))  # True True
print(is_english_word_list('xqzyw'),  is_english_word_dict('xqzyw'))   # False False

# Timing
start = time.time()
for w in word_list: is_english_word_list(w)
list_time = time.time() - start

start = time.time()
for w in word_list: is_english_word_dict(w)
dict_time = time.time() - start

print(f'List: {list_time:.2f}s   Dict: {dict_time:.4f}s')
print(f'Dict is roughly {list_time/dict_time:.0f}x faster')
````

**Teacher notes — Problem 5:**
- The timing result is the highlight: dict is typically 500-1000x faster.
- This is O(n) vs O(1) lookup made viscerally real.
- dict[w]=1 is arbitrary -- any value works. This leads to sets (which Python also has).
- Encourage students to run the list version and wait for it -- the wait is educational!

## Problem 6 — Reverse Words

**Concepts:** dict as fast lookup, string reversal, deduplication

A word is a *reverse word* if its reverse is also a valid English word.
For example: `'stressed'` reversed is `'desserts'`.

1. Write `reverse(word)` that returns the string reversed.
2. Load all words from `words.txt` into a dictionary for fast lookup.
3. Find and print all pairs. Print each pair only once (not both `(a,b)` and `(b,a)`).
4. Print how many such pairs exist.

*Note: a word that is its own reverse (palindrome) is a special case.*

````python
# Your solution here
````

````python
# === SAMPLE SOLUTION ===

def reverse(word):
    return ''.join(reversed(word))

words = open('words.txt').read().split('\n')
word_dict = {}
for w in words:
    word_dict[w] = 1

seen  = {}
pairs = []
for word in words:
    rev = reverse(word)
    if rev in word_dict and rev != word and rev not in seen:
        pairs.append((word, rev))
    seen[word] = 1

for pair in pairs[:10]:
    print(pair)

print(f'Total reverse-word pairs: {len(pairs)}')
````

**Teacher notes — Problem 6:**
- The `seen` dict elegantly avoids printing both (a,b) and (b,a).
- Fun examples to motivate the class: 'star'/'rats', 'live'/'evil', 'doom'/'mood'.
- Expected: roughly 440 unique pairs (varies by word list version).
- Dictionary comprehension `{w: 1 for w in words}` is a nice shorthand to introduce here.

## Problem 7 — Letter Frequency

**Concepts:** counting pattern, dict-to-sorted-list conversion

1. Write `count_letters(s)` that returns a dict mapping each character in `s` to its count.
   Example: `count_letters('banana')` -> `{'b': 1, 'a': 3, 'n': 2}`

2. Use it to count every letter across all words in `words.txt` (ignore spaces and newlines).

3. Sort results from most to least frequent and print:  
   `e: 106752`  
   `s:  86526`  
   `...`

4. Which 5 letters are most common? Does this match what you know about English?

````python
# Your solution here
````

````python
# === SAMPLE SOLUTION ===

def count_letters(s):
    freq = {}
    for c in s:
        if c in freq:
            freq[c] += 1
        else:
            freq[c] = 1
    return freq

print(count_letters('banana'))   # {'b': 1, 'a': 3, 'n': 2}

# Count across words.txt
words_text = open('words.txt').read()
letter_count = {}
for c in words_text:
    if c not in ' \n':
        if c in letter_count:
            letter_count[c] += 1
        else:
            letter_count[c] = 1

# Sort: [count, letter] pairs, descending
pairs = []
for letter in letter_count:
    pairs.append([letter_count[letter], letter])
pairs.sort()
pairs.reverse()

for count, letter in pairs:
    print(f'{letter}: {count:>7}')

print('Top 5:', [letter for count, letter in pairs[:5]])
````

**Teacher notes — Problem 7:**
- `count_letters` is the central pattern -- students should write it from memory by now.
- Sorting trick: put [count, letter] so .sort() orders by count automatically.
- Expected top 5: e, s, i, a, r -- matches standard English letter frequency tables.
- Extension idea: ASCII bar chart of the top 10 letters using '*' characters.

## Problem 8 (Challenge) — Inverted Index

**Concepts:** dict of lists, reversing a mapping

An *inverted index* reverses a mapping. If `home_city` maps students to cities,
the inverted index maps each city to the list of students from there:

```python
home_city = {'Alice': 'Vancouver', 'Bob': 'Victoria',
             'Carol': 'Vancouver', 'Dave': 'Kelowna', 'Eve': 'Victoria'}

# Expected inverted result:
# {'Vancouver': ['Alice', 'Carol'], 'Victoria': ['Bob', 'Eve'], 'Kelowna': ['Dave']}
```

1. Write `invert(d)` that returns the inverted index of dictionary `d`.
2. Test it on `home_city`.
3. Use your function to group words from a short list by their first letter.
4. *Stretch:* Invert the `tally` from Problem 4 to map each vote count to the list of candidates with that count.

````python
# Your solution here
home_city = {
    'Alice': 'Vancouver', 'Bob': 'Victoria',
    'Carol': 'Vancouver', 'Dave': 'Kelowna', 'Eve': 'Victoria',
}
````

````python
# === SAMPLE SOLUTION ===
home_city = {
    'Alice': 'Vancouver', 'Bob': 'Victoria',
    'Carol': 'Vancouver', 'Dave': 'Kelowna', 'Eve': 'Victoria',
}

def invert(d):
    result = {}
    for key in d:
        val = d[key]
        if val in result:
            result[val].append(key)
        else:
            result[val] = [key]
    return result

city_students = invert(home_city)
print(city_students)
# {'Vancouver': ['Alice', 'Carol'], 'Victoria': ['Bob', 'Eve'], 'Kelowna': ['Dave']}

# 3. Group words by first letter
words = ['apple', 'ant', 'bat', 'bee', 'cat', 'apricot', 'blueberry']
first_letter = {}
for w in words:
    letter = w[0]
    if letter in first_letter:
        first_letter[letter].append(w)
    else:
        first_letter[letter] = [w]
print(first_letter)
````

**Teacher notes — Problem 8:**
- Dicts of lists are very common in real programming. This is the hardest new concept in the set.
- Key step: `result[val] = [key]` starts a new list with one item -- students often forget the brackets.
- Real-world use: search engines use inverted indexes to map words to document IDs.
- The stretch goal reinforces that dict values can be any type, including lists of anything.

## Problem 9 (Stretch) — Anagram Groups

**Concepts:** canonical keys, combining all patterns

Two words are *anagrams* if they contain the same letters in any order.
Example: `'listen'` and `'silent'` are anagrams.

1. Write `canonical(word)` that returns the letters of `word` sorted alphabetically.
   Two words are anagrams iff their canonical forms are equal.
   Example: `canonical('listen')` -> `'eilnst'`

2. Load all words from `words.txt`. Build `anagram_groups`: a dict mapping each
   canonical form to the list of words with that form.

3. Find the largest anagram group and print its words.

4. How many groups have at least 2 members?

````python
# Your solution here
````

````python
# === SAMPLE SOLUTION ===

def canonical(word):
    return ''.join(sorted(word))

print(canonical('listen'))   # eilnst
print(canonical('silent'))   # eilnst

words = open('words.txt').read().split('\n')

anagram_groups = {}
for word in words:
    key = canonical(word)
    if key in anagram_groups:
        anagram_groups[key].append(word)
    else:
        anagram_groups[key] = [word]

# Largest group
best_key, best_count = '', 0
for key in anagram_groups:
    if len(anagram_groups[key]) > best_count:
        best_count = len(anagram_groups[key])
        best_key   = key

print(f'Largest group ({best_count} words): {anagram_groups[best_key]}')

# Groups with 2+ members
big = sum(1 for key in anagram_groups if len(anagram_groups[key]) >= 2)
print(f'Groups with 2+ anagrams: {big}')
````

**Teacher notes — Problem 9:**
- True stretch problem -- only for students who finished 1-8 quickly.
- Core trick: sorted string as a canonical key. Beautiful and efficient.
- This is the kind of problem where a dict is the only practical data structure.
- Expected: largest group often has 10-13 words (e.g. 'aelrst' -> alerts, alters, slater...).
- Discussion: could you solve this without a dict? O(n^2) comparisons -- extremely slow.
