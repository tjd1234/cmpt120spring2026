## Problem 2 — Trim the Middle
*Features: indexing, slicing, concatenation*

**Goal:** Shorten a long string to just its beginning and end, with `'...'` in between.

Write a function called `trim_middle` that takes a string `s` and returns a new
string containing just the **first character**, then `'...'`, then the **last character**.

| Call | Expected result |
|------|----------------|
| `trim_middle('banana')` | `'b...a'` |
| `trim_middle('hello world')` | `'h...d'` |
| `trim_middle('ab')` | `'a...b'` |

> **Assume** the input always has at least 2 characters.

````python
def trim_middle(s):
    return s[0] + '...' + s[len(s)-1]

print(trim_middle('banana'))        # should print: b...a
print(trim_middle('hello world'))   # should print: h...d
````

## Problem 4 — Does It Start With?
*Features: function, for loop, `len()`, comparison, `return`*

**Goal:** Check whether one string begins with another — without using any string methods.

Write a function called `starts_with` that takes two strings, `s` and `prefix`,
and returns `True` if `s` starts with `prefix`, and `False` otherwise.

| Call | Expected result |
|------|----------------|
| `starts_with('banana', 'ban')` | `True` |
| `starts_with('banana', 'nan')` | `False` |
| `starts_with('hi', 'hello')` | `False` (prefix is longer!) |
| `starts_with('Python', '')` | `True` (empty prefix always matches) |

> **Hints:**
> - If `prefix` is longer than `s`, the answer is immediately `False`.
> - Otherwise, compare `s[0]` with `prefix[0]`, `s[1]` with `prefix[1]`, and so on.
> - As soon as you find a mismatch, return `False`. If you get through them all, return `True`.

````python
def starts_with(s, prefix):
    for i in range(len(prefix)):
        if s[i] != prefix[i]:
            return False
    return True
    # return s[:len(prefix)] == prefix

print(starts_with('banana', 'ban'))   # True
print(starts_with('banana', 'nan'))   # False
print(starts_with('hi', 'hello'))     # False
````

## Problem 5 — Add a Border to Each Line
*Features: for loop, indexing, escape characters (`\n`), string concatenation*

**Goal:** Take a multi-line string and add `'| '` to the beginning of every line.

Write a function called `add_border` that takes a string `s` (which may contain
newline characters `'\n'`) and returns a new string with `'| '` added to the start
of each line.

**Example:**
```python
poem = 'Roses are red\nViolets are blue\nPython is fun\nAnd so are you'
print(add_border(poem))
```
```
| Roses are red
| Violets are blue
| Python is fun
| And so are you
```

> **Hints:**
> - Start with `result = '| '` (the border for the very first line).
> - Go through `s` character by character.
> - Whenever you see `'\n'`, add `'\n| '` to `result` (end the line and start the next border).
> - For any other character, just add it to `result` normally.

````python
def add_border(s):
    result = '| '
    for c in s:
        result += c
        if c == '\n':
            result += '| '
    return result
    # return '| ' + s.replace('\n', '\n| ')

poem = 'Roses are red\nViolets are blue\nPython is fun\nAnd so are you'
print(add_border(poem))
````

## Problem 6 — Count the Vowels
*Features: function, for loop, comparison, counting, `return`*

**Goal:** Count how many vowels are in a string.

Write a function called `count_vowels` that takes a string `s` and returns the
number of vowel characters in it. Count both lowercase and uppercase vowels
(a, e, i, o, u, A, E, I, O, U).

| Call | Expected result |
|------|----------------|
| `count_vowels('hello')` | `2` |
| `count_vowels('Python')` | `1` |
| `count_vowels('rhythm')` | `0` |
| `count_vowels('AEIOU')` | `5` |

> **Hint:** Keep a counter starting at `0`. For each character, check whether it
> equals `'a'`, or `'e'`, or `'i'`, … (10 comparisons total, connected with `or`).

````python
def count_vowels(s):
    result = 0
    for c in s:
        if c.lower() in 'aeiou':
            result += 1
    return result

print(count_vowels('hello'))   # should print: 2
print(count_vowels('Python'))  # should print: 1
````

## Problem 7 — Reverse a String
*Features: function, for loop, string concatenation, `return`*

**Goal:** Return a string with all its characters in reverse order.

Write a function called `reverse` that takes a string `s` and returns a new string
with the same characters in the opposite order. **Do not** use Python's built-in
reversing features.

| Call | Expected result |
|------|----------------|
| `reverse('hello')` | `'olleh'` |
| `reverse('Python')` | `'nohtyP'` |
| `reverse('racecar')` | `'racecar'` |
| `reverse('')` | `''` |

> **Hint:** Walk *backwards* through `s` using `range(len(s)-1, -1, -1)`.
>
> **Bonus challenge:** Can you write a one-line version using slicing? What does `s[::-1]` do?

````python
def reverse(s):
    result = ''
    for c in s:
        result = c + result
    return result

print(reverse('hello'))   # should print: olleh
print(reverse('Python'))  # should print: nohtyP
````

## Problem 8 — Is It a Palindrome?
*Features: function, for loop, indexing, comparison, `return`*

**Goal:** Check whether a string reads the same forwards and backwards.

A **palindrome** is a word or phrase that is identical when reversed — like
`'racecar'`, `'level'`, or `'madam'`.

Write a function called `is_palindrome` that takes a string `s` and returns
`True` if it is a palindrome, and `False` if it is not.

| Call | Expected result |
|------|----------------|
| `is_palindrome('racecar')` | `True` |
| `is_palindrome('level')` | `True` |
| `is_palindrome('hello')` | `False` |
| `is_palindrome('a')` | `True` |
| `is_palindrome('')` | `True` |

> **Hints:**
> - Use a loop that runs `len(s) // 2` times.
> - At step `i`, compare `s[i]` with `s[-(i+1)]`.
> - If any pair doesn't match, return `False` immediately.
>
> **Bonus:** Can you write a one-liner using `reverse` from Problem 7?

````python
def is_palindrome(s):
    # return s[::-1] == s
    for i in range(len(s)):
        # s[0] == s[-1]
        # s[1] == s[-2]
        # s[2] == s[-3]
        # ...
        if s[i] != s[-(i+1)]:
            return False
    return True


print(is_palindrome('racecar'))   # True
print(is_palindrome('hello'))     # False
print(is_palindrome('a'))
print(is_palindrome(''))
print(is_palindrome('aa'))
````
