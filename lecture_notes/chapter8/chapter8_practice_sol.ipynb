````python
def starts_with(s, prefix):
    """Returns True if s starts with prefix, False otherwise.

    starts_with('banana', 'ban')   -> True
    starts_with('banana', 'nan')   -> False
    starts_with('hi', 'hello')     -> False
    starts_with('Python', '')      -> True
    """
    # A prefix longer than the string can never match
    if len(prefix) > len(s):
        return False

    for i in range(len(prefix)):
        if s[i] != prefix[i]:
            return False   # mismatch found — stop immediately

    return True   # survived all comparisons

# --- Tests ---
print(starts_with('banana', 'ban'))    # True
print(starts_with('banana', 'nan'))    # False
print(starts_with('hi', 'hello'))      # False
print(starts_with('Python', ''))       # True
print(starts_with('abc', 'abc'))       # True  (exact match)
````

**Teaching notes — Problem 4:**
- The early-return pattern (`return False` inside the loop) is important — we don't
  need a `found` flag, we just return right away.
- The empty-prefix case is handled for free: `range(0)` produces no iterations,
  so the loop never runs and we fall through to `return True`.
- Ask: what would happen without the `len(prefix) > len(s)` guard?
  → `IndexError` when we try `s[i]` for an out-of-range index.

````python
def add_border(s):
    """Returns s with '| ' prepended to every line.

    add_border('Hello\nWorld') -> '| Hello\n| World'
    """
    result = '| '          # border for the very first line

    for i in range(len(s)):
        if s[i] == '\n':
            result = result + '\n| '   # end line, start bordered next line
        else:
            result = result + s[i]     # ordinary character — keep as-is

    return result

# --- Tests ---
poem = 'Roses are red\nViolets are blue\nPython is fun\nAnd so are you'
print(add_border(poem))
print()
print(add_border('Hello\nWorld'))
print()
print(add_border('single line'))   # no newlines — border only at start
````

````python
def reverse(s):
    """Returns s with its characters reversed.

    reverse('hello')   -> 'olleh'
    reverse('Python')  -> 'nohtyP'
    reverse('racecar') -> 'racecar'
    reverse('')        -> ''
    """
    result = ''
    for i in range(len(s) - 1, -1, -1):
        result = result + s[i]
    return result

# --- Tests ---
print(reverse('hello'))     # olleh
print(reverse('Python'))    # nohtyP
print(reverse('racecar'))   # racecar
print(reverse(''))          # (empty string)
print(reverse('a'))         # a

# Bonus — one-liner using slice notation:
def reverse_slice(s):
    return s[::-1]    # step of -1 means "go backwards"

print(reverse_slice('hello'))   # olleh
````

> **Hints:**

> - At step `i`, compare `s[i]` with `s[-(i+1)]`.
> - If any pair doesn't match, return `False` immediately.

````python
def is_palindrome(s):
    """Returns True if s is a palindrome, False otherwise.

    is_palindrome('racecar') -> True
    is_palindrome('level')   -> True
    is_palindrome('hello')   -> False
    is_palindrome('a')       -> True
    is_palindrome('')        -> True
    """
    for i in range(len(s) // 2):
        if s[i] != s[-(i + 1)]:
            return False
    return True

# --- Tests ---
print(is_palindrome('racecar'))   # True
print(is_palindrome('level'))     # True
print(is_palindrome('hello'))     # False
print(is_palindrome('a'))         # True
print(is_palindrome(''))          # True
print(is_palindrome('ab'))        # False

# Bonus — using reverse from Problem 7:
def is_palindrome_v2(s):
    return s == reverse(s)   # requires reverse() to be defined above

# Bonus — recursive version (preview for Problem 11 stretch goal):
def is_palindrome_r(s):
    if len(s) <= 1:                   # base case: 0 or 1 chars are always palindromes
        return True
    if s[0] != s[-1]:                 # first and last don't match
        return False
    return is_palindrome_r(s[1:-1])   # check the middle part
````

**Teaching notes — Problem 8:**
- Why `len(s) // 2`? For a 6-char string we only need 3 pairs; for a 7-char string
  the middle character needs no partner. Integer division handles both cases.
- Unpack `s[-(i+1)]` on the board:
  - `i=0` → `s[-1]` (last), `i=1` → `s[-2]` (second-to-last), and so on.
- The one-liner `s == reverse(s)` is elegant — shows how earlier functions compose.
- For the recursive version, trace `is_palindrome_r('racecar')` step by step:
  `'racecar'` → `'aceca'` → `'cec'` → `'e'` → `True`.

````python
# ── Helper functions ──────────────────────────────────────────────────────────

def count_vowels(s):
    """Returns the number of vowels in s."""
    count = 0
    for i in range(len(s)):
        ch = s[i]
        if (ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u' or
                ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U'):
            count = count + 1
    return count

def count_spaces(s):
    """Returns the number of space characters in s."""
    count = 0
    for i in range(len(s)):
        if s[i] == ' ':
            count = count + 1
    return count

def reverse(s):
    """Returns s with its characters reversed."""
    result = ''
    for i in range(len(s) - 1, -1, -1):
        result = result + s[i]
    return result

def is_palindrome(s):
    """Returns True if s reads the same forwards and backwards."""
    for i in range(len(s) // 2):
        if s[i] != s[-(i + 1)]:
            return False
    return True

# ── Main report function ───────────────────────────────────────────────────────

def make_report(s):
    """Builds and returns a formatted analysis report for string s."""
    n = len(s)

    # First/last 5 characters (use the whole string if it's shorter than 5)
    if n <= 5:
        first5 = s
        last5  = s
    else:
        first5 = s[:5]
        last5  = s[-5:]

    if is_palindrome(s):
        palindrome_label = 'Yes!'
    else:
        palindrome_label = 'Nope.'

    report  = '=== String Report ===\n'
    report  = report + 'String:      ' + s                      + '\n'
    report  = report + 'Length:      ' + str(n)                 + '\n'
    report  = report + 'Vowels:      ' + str(count_vowels(s))   + '\n'
    report  = report + 'Spaces:      ' + str(count_spaces(s))   + '\n'
    report  = report + 'First 5:     ' + first5                 + '\n'
    report  = report + 'Last 5:      ' + last5                  + '\n'
    report  = report + 'Reversed:    ' + reverse(s)             + '\n'
    report  = report + 'Palindrome?  ' + palindrome_label       + '\n'

    return report

# ── Run it ────────────────────────────────────────────────────────────────────

for s in ['racecar', 'Hello World']:
    report = make_report(s)
    print(report)

    report_file = open('string_report.txt', 'w')
    report_file.write(report)
    report_file.close()

print('Report saved to string_report.txt!')
````

````python
# Bonus — recursive is_palindrome
def is_palindrome_r(s):
    """Recursive palindrome check."""
    if len(s) <= 1:                   # base case: 0 or 1 characters
        return True
    if s[0] != s[-1]:                 # first and last don't match
        return False
    return is_palindrome_r(s[1:-1])   # check the middle

# Trace for 'racecar':
# is_palindrome_r('racecar') -> r==r, recurse on 'aceca'
# is_palindrome_r('aceca')   -> a==a, recurse on 'cec'
# is_palindrome_r('cec')     -> c==c, recurse on 'e'
# is_palindrome_r('e')       -> len 1 -> True ✓

print(is_palindrome_r('racecar'))   # True
print(is_palindrome_r('hello'))     # False
print(is_palindrome_r(''))          # True
````
