# ROT13: A Text Encryption Algorithm

**Why does no talk about peanut butter before it's spreadable?**

**Orpnhfr gung jbhyq or ahgf.**

If you're not sure how to interpret that punchline, then you are in luck. It's
encrypted in [ROT13](https://en.wikipedia.org/wiki/ROT13), a simple encryption
algorithm that replaces each letter in a text with the letter 13 letters after
it in the alphabet (wrapping around to the beginning of the alphabet if
necessary).

Letters translate as follows:

```
       Original: ABCDEFGHIJKLMNOPQRSTUVWXYZ abcdefghijklmnopqrstuvwxyz
ROT13 encooding: NOPQRSTUVWXYZABCDEFGHIJKLM nopqrstuvwxyzabcdefghijklm
```

The letter `A` is replaced with `N`, `k` with `x`, and so on. Characters that
are not letters are not changed. 

For example, the string `Hello, World!` is `Uryyb, Jbeyq!` in ROT13.

ROT13 is *not* a secure encryption algorithm: it is easy to recognize and decode
by hand. In the past it was used on
[Usenet newsgroups](https://en.wikipedia.org/wiki/Usenet) to obscure things like
joke punchlines, or to hide spoilers for movies or games.

# The `ord` and `chr` functions

Every character has a standard code
[Unicode](https://en.wikipedia.org/wiki/Unicode) value, and Python's `ord`
function returns that value. For example:

````python
print(ord('A')) # 65
print(ord('B')) # 66
print(ord('C')) # 67
print(ord('a')) # 97
print(ord('b')) # 98
print(ord('c')) # 99
````

If you have a code value, you can convert it back to a character with `chr`:

````python
print(chr(65)) # A
print(chr(66)) # B
print(chr(67)) # C
print(chr(97)) # a
print(chr(98)) # b
print(chr(99)) # c
````

We can use this to write functions to test for letters:

````python
#  TODO: implement is_lower, is_upper, is_letter
def is_upper(c):
    # return len(c) == 1 and 65 <= ord(c) <= 92
    # return len(c) == 1 and ord('A') <= ord(c) <= ord('Z')
    return len(c) == 1 and 'A' <= c <= 'Z'

def is_lower(c):
    return len(c) == 1 and 97 <= ord(c) <= 122

def is_letter(c):
    return is_lower(c) or is_upper(c)

print(is_letter('A')) # True
print(is_letter('b')) # True
print(is_letter('q')) # True
print(is_letter('x')) # True
print(is_letter('!')) # False
print(is_letter('ab')) # False
````

We can also write a function that converts characters to ROT13. Let's start by
considering just the upper case letters.

As a first attempt, we could just add 13 to the code value:

```
A 65 + 13 = 78 N
B 66 + 13 = 79 O
C 67 + 13 = 80 P
D 68 + 13 = 81 Q
E 69 + 13 = 82 R
F 70 + 13 = 83 S
G 71 + 13 = 84 T

J 74 + 13 = 87 W

N 78 + 13 = 91 [   # wrong, should be 65 A

Q 81 + 13 = 94 ^   # wrong, should be 68 D
R 82 + 13 = 95 _   # wrong, should be 69 E
S 83 + 13 = 96 `   # wrong, should be 70 F
T 84 + 13 = 97 a   # wrong, should be 71 G
U 85 + 13 = 98 b   # wrong, should be 72 H
V 86 + 13 = 99 c   # wrong, should be 73 I

Y 89 + 13 = 102 f  # wrong, should be 76 L
Z 90 + 13 = 103 g  # wrong, should be 77 M
```

Adding 13 works for `A` to `M`, but fails for `N` to `Z`. When we get to `N` the
values should "wrap around" back to 65 for `A`. For `N`, the correct result is
78, for `O` it's 79, and so on.

Looking at the code values for the wrong letters, we note that the ROT13 code is
13 less than it's code. E.g. `N` is 78, we want `A` 65, and `78 - 13 = 65`. Similarly,
`Y` is 89, we want `L` 76, and `89 - 13 = 76`. So for `N` to `Z`, we *subtract* 13 to get its ROT13 code.

So there are two rules for converting upper case letters to ROT13:

- For `A` to `M`, add 13.
- For `N` to `Z`, subtract 13.


````python
# TODO: implement rot13_upper(c) that converts one upper case character to ROT13
# (no upper case characters are returned unchanged)

# def rot13_upper(c):
#     if is_upper(c):
#         code = ord(c)
#         if 'A' <= c <= 'M':
#             return chr(code + 13)
#         elif 'N' <= c <= 'Z':
#             return chr(code - 13)
#     else: # c is not upper case
#         return c

def rot13_upper(c):
    if 'A' <= c <= 'M':
        return chr(ord(c) + 13)
    elif 'N' <= c <= 'Z':
        return chr(ord(c) - 13)
    else:
        return c

print(rot13_upper('A')) # N
print(rot13_upper('Z')) # M
print(rot13_upper('M')) # Z
print(rot13_upper('T')) # G
print(rot13_upper('5'))
````

For lower case letters, the rules are similar:

- For `a` to `m`, add 13.
- For `n` to `z`, subtract 13.

````python
# TODO: write rot13_lower that converts lower cases letters to ROT13

def rot13_lower(c):
    if 'a' <= c <= 'm':
        return chr(ord(c) + 13)
    elif 'n' <= c <= 'z':
        return chr(ord(c) - 13)
    else:
        return c

print(rot13_lower('a')) # n
print(rot13_lower('z')) # m
print(rot13_lower('m')) # z
print(rot13_lower('t')) # g
````

Combining these two functions give us a complete ROT13 function:

````python
# TODO: return the ROT13 of a lower case or upper case letter

def rot13_letter(c):
    if is_lower(c):
        return rot13_lower(c)
    elif is_upper(c):
        return rot13_upper(c)
    else:
        return c

def rot13(text):
    result = ''
    for c in text:
        result += rot13_letter(c)
    return result

print(rot13('Hello, World!')) # Uryyb, Jbeyq!
print(rot13('Uryyb, Jbeyq!')) # Hello, World!

# Why does no talk about peanut butter before it's spreadable?

print(rot13('Orpnhfr gung jbhyq or ahgf.'))
````

A nice of feature of `rot13` is that it is its own inverse. If you apply it
twice you get back the original text:

````python
print(rot13('Once upon a time'))
print(rot13(rot13('Once upon a time'))) # Once upon a time
````

So `rot13` can both encode and decode text.

## Encoding a File

Suppose you want to encode an entire text file into ROT13. You could do it like this:

- Open the input file for reading.
- Open a new output file for writing the result.
- Read the input file one line at a time.
- Convert each line to ROT13.
- Write the result to the output file.
- Close both files.

Here's a function that does this:

````python
# TODO: implement the about algorithm

def rot13_file(input_file, output_file):
    infile = open(input_file, 'r')
    outfile = open(output_file, 'w')
    
    for line in infile:
        outfile.write(rot13(line))

    infile.close()
    outfile.close()

# print the original file

rot13_file('secret.txt', 'decoded.txt')

# print the new file
````

Notice that you can read a text file as a string with `open(filename).read()`.
